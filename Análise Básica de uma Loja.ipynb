```python
import pandas as pd
```

```python
tabela = pd.read_excel('/content/drive/MyDrive/Analista de Dados (1)/analise 2/Vendas.xlsx')
```

```python
tabela
```

```python
#vendo quantas linhas e colunas existem na tabela
tabela.shape
```

```python
#vendo todas as colunas
tabela.columns
```

```python
#ver o faturamento da empresa - geral - somar todos os valores da coluna final
#use seu dataframe tabela, atribua a coluna que quer somar e use o comando .sum()
#observe que foi criado outra variavel chamado faturamento_total

faturamento_total = tabela['Valor Final'].sum()
```

```python
#valor do faturamento - soma de todos os valores da ultima coluna
#se pergunte de onde vem esse valor / qual loja está vendendo mais / qual produto está vendendo mais
faturamento_total
```

```python
#calcular o faturamento por loja - use a coluna id loja e valor final
#quando voce quer passar mais de uma coluna, use dois colchetes.
faturamento_por_loja = tabela[['ID Loja' , 'Valor Final']]
```

```python
#aqui esta o faturamento de cada loja
#observe que aqui ainda nao esta o faturamento por loja pois, se observar existe 
#repetição de loja. o objetivo é soma-las - podendo agrupar
faturamento_por_loja
```

```python
#use o comando groupby(' aqui o nome da coluna que quer agrupar')
#aqui voce está dizendo para agrupar as colunas repetidas e somar( as que restam tambem)
#groupby é o comando para voce agrupar as colunas, passe dentro dos parenteses os paramentros
faturamento_por_loja = tabela[['ID Loja' , 'Valor Final']].groupby('ID Loja').sum()
```

```python
#aqui esta o faturamento de cada loja
#observe o que ha de diferente na loja que está com mais valor
faturamento_por_loja
```

```python
#subtende-se que a loja com maior valor tem produtos que as outras nao tem
# aqui agrupei a coluna loja + a coluna produto e pedi o somatorio
#assim foi feito para mostrar os produtos de cada loja e seus valores
faturamento_por_produto = tabela[['ID Loja' , 'Produto' , 'Valor Final']].groupby(['ID Loja' , 'Produto']).sum()
```

```python
#observe que nas outras lojas nao tem o tipo de "bermuda liso" , item que mais vende na loja de campinas
faturamento_por_produto
```
